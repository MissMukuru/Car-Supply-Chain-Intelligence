```python
#Multiplying the sales data by the discount and adding the results

import numpy as np
import pandas as pd
Chain = pd.read_csv('Car_data.csv')
column_to_convert = 'Sales'
col_to_convert = 'Discount'
sales_array = Chain[column_to_convert].values
Discount_array = Chain[col_to_convert].values

print(sales_array)
print(Discount_array)

Discounted_price = np.multiply(sales_array, Discount_array)
print(Discount_array)

new_sales = np.prod(Discounted_price)
print(new_sales)
```
